```python
import pandas as pd
import scipy.stats as st
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def teste(row):
    return row['Inicio(seg)'] + row['Duracao(seg)']
```

```python
df = pd.read_table("../Saidas/Teste1/iperfs.txt", sep="\t")
df["Fim"] = df.apply (teste, axis=1)
df.drop(['Duracao(seg)','PCli', 'PServ','Banda(Kbps)'], axis=1, inplace=True)
df
```

```python
# Função para cálculo do intervalo de confiança de 95%
def confinterval (data):
    confidence = 0.95
    return st.sem (data) * st.t.ppf ((1 + confidence) / 2.0, len (data) - 1);
```
